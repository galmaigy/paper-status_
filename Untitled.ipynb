```python
import folium
from IPython.display import display
import pandas as pd
import numpy as np
```

```python
df_hour = pd.read_excel('data/days.xlsx',  dtype={'leftover': int, 'cum':int,'estimated':int})
display(df_hour)
```

```python
df_copy = df_hour.copy()

mask = df_copy.PharmacyID.isin(['1PITT','3INDI'])
df_map = df_copy.loc[mask,:]
#df_map=df_map[df_map.PharmacyID.isin(['1PITT','3INDI','4KCTY'])]

display(df_map)
```

```python
# base map creating
def generateBaseMap(default_location=[39.9612 , -82.9988], default_zoom_start=6):
    base_map = folium.Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)
    return base_map

base_map = generateBaseMap()

base_map
```

```python
cum_est =[]

cum_est.append(df_map[['latitude','longitude','estimated']].values.tolist())
cum_est
```

```python
import folium.plugins as plugins

hm=plugins.HeatMapWithTime(cum_est,
                           radius=5, 
                           gradient={0.2: 'blue', 0.4: 'lime', 0.6: 'orange', 1: 'red'}, 
                           min_opacity=0.5, max_opacity=0.8, use_local_extrema=True)

hm.add_to(base_map)

base_map
```

```python
est_list = []
est_list.append(df_map[['latitude','longitude','estimated']].values.tolist())
est_list
```

```python
import folium.plugins as plugins

em=plugins.HeatMapWithTime(est_list)

em.add_to(base_map)

base_map.save('time_est.html')
```
