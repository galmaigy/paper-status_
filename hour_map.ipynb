```python
import folium
from IPython.display import display
import pandas as pd
import numpy as np
```

```python
df_hour = pd.read_excel('data/hour.xlsx',  dtype={'leftover': int, 'cum':int})
display(df_hour)
```

```python
df_copy = df_hour.copy()

mask = df_copy.PharmacyID.isin(['1PITT','3INDI'])
df_map = df_copy.loc[mask,:]
#df_map=df_map[df_map.PharmacyID.isin(['1PITT','3INDI','4KCTY'])]

display(df_map)
```

```python
base_map = folium.Map(location=[39.9612 , -82.9988],
                        zoom_start=6,
                        tiles='Stamen Toner')

base_map
```

```python
cum_list = df_map[['latitude','longitude','cum']].values.tolist()
cum_list
```

```python
from folium.plugins import HeatMapWithTime

m=HeatMapWithTime(cum_list, radius=5, gradient={500: 'blue', 1500: 'lime', 2500: 'orange', 3500: 'red'}, min_opacity=0.5, max_opacity=0.8, use_local_extrema=True).add_to(base_map)
m.save('HeatMapWithTime_hour_1.html')

```
