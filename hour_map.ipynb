```python
import folium
from IPython.display import display
import pandas as pd
import numpy as np
```

```python
df_hour = pd.read_excel('data/hour.xlsx',  dtype={'leftover': int, 'cum':int, 'estimated':int})
display(df_hour)
```

```python
df_copy = df_hour.copy()

mask = df_copy.PharmacyID.isin(['1PITT','3INDI'])
df_map = df_copy.loc[mask,:]
#df_map=df_map[df_map.PharmacyID.isin(['1PITT','3INDI','4KCTY'])]

display(df_map)
```

```python
base_map = folium.Map(location=[39.9612 , -82.9988],
                        zoom_start=6,
                        tiles='Stamen Toner')

base_map
```

```python
cum_list2 =[]

cum_list2.append(df_map[['latitude','longitude','cum']].values.tolist())
cum_list2

```

```python
est_list = []
est_list.append(df_map[['latitude','longitude','estimated']].values.tolist())
est_list
```

```python
from folium.plugins import HeatMapWithTime

m=HeatMapWithTime(cum_list2, radius=5, gradient={0.2: 'blue', 0.4: 'lime', 0.6: 'orange', 1: 'red'}, min_opacity=0.5, max_opacity=0.8, use_local_extrema=True).add_to(base_map)
m.save('HeatMapWithTime_hour_3.html')

```

```python
import folium.plugins as plugins

hm=plugins.HeatMapWithTime(cum_list2)

hm.add_to(base_map)

base_map
```

```python
import folium.plugins as plugins

hm=plugins.HeatMapWithTime(cum_list2,
                           radius=5,
                           gradient={0.25: 'blue', 0.5: 'lime', 0.75: 'orange', 1: 'red'}, 
                           min_opacity=0.5, 
                           max_opacity=0.8, 
                           use_local_extrema=True
                          )

hm.add_to(base_map)

base_map.save('newHeatMapWithTime.html')
```

```python
import folium.plugins as plugins

em=plugins.HeatMapWithTime(est_list, 
                           radius=5, 
                           gradient={0.2: 'blue', 0.4: 'lime', 0.6: 'orange', 1: 'red'}, 
                           min_opacity=0.5, max_opacity=0.8, use_local_extrema=True)

em.add_to(base_map)
base_map.save('est_1.html')
```
