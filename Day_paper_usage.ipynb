```python
import folium
from IPython.display import display
import pandas as pd
import numpy as np
```

```python
df_day = pd.read_excel('data/day.xlsx', index_col = 'name', dtype={'estimated': int})
display(df_day)
```

```python
df_day.info()
```

```python
df_map = df_day.copy()


base_map = folium.Map(location=[37.0902 , -95.7129],
                        zoom_start=5,
                        tiles='Stamen Toner')

base_map

base_map.save('map/base_map.html')
```

```python
def color(est): 
    if est in range(0,400): 
        col = 'green'
    elif est in range(401,800): 
        col = 'yellow'
    elif est in range(801,1200): 
        col = 'orange'
    else: 
        col='red'
    return col 

for name, lat, lng, est in zip(df_map.index, df_map.latitude, df_map.longitude, df_map.estimated):
    folium.CircleMarker(location = [lat,lng],
                        radius = 5, 
                        color = color(est),
                        fill =True, 
#                         fill_color ='coral',
                        fill_opacity =0.7,
                        popup= [name,est] 
                       ).add_to(base_map)

# folium.Marker([40.3675689, -80.1093305], popup='HUB',icon=folium.Icon(color='blue', icon='home')).add_to(pa_map_)

display(base_map)
base_map.save('map/day.html')
```
